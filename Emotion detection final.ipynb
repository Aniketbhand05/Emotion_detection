```python
import cv2 as cv
from tensorflow import keras
from keras.models import load_model
from PIL import Image
import numpy as np
from keras.preprocessing.image import img_to_array

model = load_model("emotion_trial_1.hdf5")
classifier = cv.CascadeClassifier("haarcascade_frontalface_default.xml")


emotion_label = ['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise']

cam = cv.VideoCapture(0)
while True :
    _,frame = cam.read()
    frame = cv.flip(frame , 1)
    gray = cv.cvtColor(frame , cv.COLOR_RGB2GRAY)
    face = classifier.detectMultiScale(gray , 1.3, 10)
    for x,y,w,h in face:
        
        cv.rectangle(frame , (x,y) , (x+w , y+h) , (0,255,0) , 2)
        roi_gray = gray[y:y+h , x:x+w]
        roi_gray = cv.resize(roi_gray , (48,48) , interpolation = cv.INTER_AREA)
        
        
        if(np.sum([roi_gray]) !=0) :
#             roi = roi_gray/255
#             roi = img_to_array(roi)
            
            roi = np.expand_dims(roi_gray,axis=0)
            roi = np.tile(roi[..., None], (1, 1, 3))
            print(roi.shape)
            predictions = model.predict(roi)[0]
            label = emotion_label[np.argmax(predictions)]
            label_pos = (x,y)
            cv.putText(frame , label , label_pos ,cv.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2)
        else:
            cv.putText(frame,'No Faces',(30,80),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2)
    cv.imshow('Emotion Detector',frame)
    
    if (cv.waitKey(1) == 27):
        
        break
        
cam.release()
cv.destroyAllWindows()
```
